```python
import os
import pandas as pd 

PATH = '../../data/dogfacenet/aligned/'
df = pd.read_csv(PATH+'labels/via_region_data(5).csv')
df.head()
```

```python
df2 = df[df.region_count == 3]
```

```python
len(df2)/3
```

```python
df3 = df2[['filename','region_shape_attributes']]
```

```python
df3.head()
```

```python
from ast import literal_eval
df3['region_shape_attributes'] = df3['region_shape_attributes'].apply(literal_eval)
```

```python
df3.head()
```

```python
df3['id'] = [i for i in range(len(df3))]
```

```python
df3 = df3.set_index('id')
```

```python
dic = df3['region_shape_attributes'][0]
dic
```

```python
dic['cx']
```

```python
df4 = pd.DataFrame(index=[i for i in range(len(df3)//3)],columns=['filename','lex','ley','rex','rey','nox','noy'])
```

```python
df4['filename']=[df3['filename'][i] for i in range(0,len(df3),3)]
df4['lex'] = [df3['region_shape_attributes'][i]['cx'] for i in range(0,len(df3),3)]
df4['ley'] = [df3['region_shape_attributes'][i]['cy'] for i in range(0,len(df3),3)]
df4['rex'] = [df3['region_shape_attributes'][i]['cx'] for i in range(1,len(df3),3)]
df4['rey'] = [df3['region_shape_attributes'][i]['cy'] for i in range(1,len(df3),3)]
df4['nox'] = [df3['region_shape_attributes'][i]['cx'] for i in range(2,len(df3),3)]
df4['noy'] = [df3['region_shape_attributes'][i]['cy'] for i in range(2,len(df3),3)]
```

```python
df4.to_csv(PATH+'labels/reformated.csv')
```
